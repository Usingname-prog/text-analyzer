```python
import os

import string

from google.colab import files as x        #allows user to upload files

if os.path.exists("read.txt"):
  names = "read.txt"
  print("File is already stored in system as:", names)     #happens only with txt file named as "read.txt" (if uploaded)

else:
  documents = x.upload()

  if len(documents) == 0:                 #shows when user declines file uploading
    print("File wasn't uploaded")
    text = None
    splitted = None
    names = None

  else:
    names = list(documents.keys())[0]
    print("File uploaded:", names)
if names is None:
  print("No files, maybe you decided not to upload them")
if names is not None:
    with open(names, "r", encoding="utf-8-sig") as doc:      #reads and closes files that were uploaded as utf-8 txt files
        text = doc.read()
    if text.strip() == "":
      print("Programm cannot see anything in this file")
    else:
      separated = len(text)
      together = len(text.replace( ' ', ''))
      cos = 0

      for ch in text:
        if ch=='.':
          cos = cos+1
        if ch=='?':
          cos = cos+1
        if ch=='!':
          cos = cos+1
      sin = 0
      sine = text.split('\n\n')
      for smth in sine:
        if smth.strip() !='':
          sin = sin+1
      print("\nText's summary:")
      print("\nSentences:", cos)
      print("\nParagraphs:", sin)
      print("\nSum of сharacters, spaces included:", separated)
      print('\nSum of characters, excluding spaces:', together)

      def prepr(text):
        lowered = text.lower()        #commonly removes punctuation so that future countings would be more precise
        extra_symb = "''""...--"
        clean_symb = string.punctuation + extra_symb
        cleaned = str.maketrans({ch: " " for ch in clean_symb})
        final_text = lowered.translate(cleaned)
        splitted = final_text.split()
        return splitted
      def frequency(splitted):
        freq = {}
        for word in splitted:
          if word in freq:
            freq[word] = freq[word] + 1
          else:
            freq[word] = 1
        return freq

      def themes(freq, outs):
        inside = {}
        for out in outs:
          inside[out] = 0
        for word in freq:
          for out in outs:
            if word in outs[out]:
              inside[out] = inside[out] + freq[word]     #counts key words of topics from dictionary (see below)
        pos_freq = 0
        pos_out = []
        for out in inside:
          if inside[out] > pos_freq:
            pos_freq = inside[out]
            pos_out = [out]
          elif inside[out] == pos_freq and pos_freq !=0:
            pos_out.append(out)
        return inside, pos_freq, pos_out

      def repeating(freq, end):
        Richard = list(freq.items())
        y = len(Richard)
        for sort in range(y):
          max_index = sort
          for sub in range(sort+1, y):
            if Richard[sub][1]> Richard[max_index][1]:   #compares frequencies between appearances of words in text
              max_index = sub                            #changes places of words if one appeared more
          old_data = Richard[sort]
          Richard[sort] = Richard[max_index]
          Richard[max_index] = old_data
        return Richard[:end]

      outs = {                        #possible topics, easy to change as these are just dictionaries
        "Science": ["research", "investigation", "observations", "innovation", "study", "experiment", "test", "analysis"],
        "Technology": ["data", "smart", "system", "algorithm", "computer", "device", "gadget", "network", "software", "hardware"],
        "Politics": ["summit", "global", "leader", "president", "government", "crisis", "policy", "country", "agreement", "law"],
        "Nature": ["mountain", "environment", "climate", "ecology", "atmosphere", "hemisphere", "natural", "animal"],
        "Economics": ["inflation", "money", "spendings", "tax", "subsidy", "stock", "market", "invest", "finance", "management"],
        "Sports": ["game", "ball", "match", "play", "team", "nike", "sneakers", "field"],
        "Healthcare": ["hospital", "care", "therapy", "pills","drugs", "patient", "accident"]
      }
      splitted = prepr(text)
      print("\nWord count:", len(splitted))
      if len(splitted) == 0:
        print("No words were found")
      else:
        freq = frequency(splitted)
        print("Non-repeating words:", len(freq))

        inside, pos_freq, pos_out = themes(freq,outs)
        print("\nTopic scores:")
        for out in inside:
          print(out + ':', inside[out])
        if pos_freq == 0:
          print('\nTheme is really hard to define')
        elif len(pos_out) == 1:
          print("\nPredicted topic:", pos_out[0])
        else:
          print('\nText covers these possible main themes:')
          print(' and '.join(pos_out))
        most_freq = repeating(freq, 13)
        places = []
        for word, count in most_freq:
         places.append("\n" '\n' + word+ ": " + str(count))
        print("\nTop words:", ' '.join(places[:5]))
        print("\nFirst 24 (but divided) words:", " ".join(splitted [:12]), '\n', ' '.join(splitted[12:24]))
```
